# Data Visualization

Data visualization is used to explore your data and to communicate your data. My goal for you isn't to make you a data viz wizard, but to give you enough of an understanding to be able to explore your data and jumpstart your own learning.  

The **matplotlib** library is great for making simple plots. If you want to make elaborate interactive visualizations, it's probably not for you. But it's good for quick and dirty data viz that is relatively customizable. 

The **seaborn** library is built on matplotlib and has a lot more "out-of-the-box" functionality for quick plots.

We will need to install Seaborn with

`conda install Seaborn`

[Instructions to install](https://seaborn.pydata.org/installing.html)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Load in the iris dataset

```python
iris = pd.read_csv("iris.csv")
```

```python
iris.head()
```

#### Pandas plotting works for simple plots

```python
iris.plot(kind="scatter", x="sepal_length", y="sepal_width")
plt.show()
```

#### Change color and size 

```python
iris.plot(kind="scatter", x="sepal_length", y="sepal_width",color="red",s=80 )
plt.show()
```

#### Histograms

```python
sns.distplot(iris.sepal_length, kde=False, rug=True)
plt.show()
```

#### Seaborn's jointplot will produce a scatterplot and histograms

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, size=5)
plt.show()
```

#### Assign each species an individual color.

```python
sns.FacetGrid(iris, hue="species", size=5) \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend()
plt.show()
```

#### Use Seaborn to make a boxplot

```python
sns.boxplot(x="species", y="petal_length", data=iris )
plt.show()
```

#### Use stripplot to add datapoints onto box plot

```python
# Insert jitter=True so that the data points remain scattered and not piled into a verticle line.
# Assign ax to each axis, so that each plot is ontop of the previous axis. 

ax= sns.boxplot(x="species", y="petal_length", data=iris)
ax= sns.stripplot(x="species", y="petal_length", data=iris, jitter=True, edgecolor="gray")
plt.show()
```

#### Seaborn's violinplots show density of data.

```python
sns.violinplot(x="species", y="petal_length", data=iris, size=6)
plt.show()
```

#### Kdeplot plots densities

```python
sns.FacetGrid(iris, hue="species", size=5) \
   .map(sns.kdeplot, "petal_length") \
   .add_legend()
plt.show()
```

#### Pairplot shows relationships between all variables

```python
sns.pairplot(iris, hue="species", size=3)
plt.show()
```

#### Heatmap

```python
sns.heatmap(iris.corr())
plt.title('Covariance plot')
plt.show()
```
