## Clean Data

Remove first 16 columns

```python
drop_cols = survey.columns[0:17].values
```

```python
survey = survey.drop(drop_cols, axis=1)
```

```python
survey.head()
```

Remove first two rows

```python
survey = survey.drop([0,1], axis = 0)
```

```python
survey.head()
```

```python
survey.tail()
```

Remove Q3. The wording was vague and unclear in the survey.

```python
survey = survey.drop('Q3', axis=1)
```

Rename columns

```python
col_names = ["ig_use", "photos_daily", "check_3_hrs", "tried_soylent", "soylent_freq", "interested", "lifestyle", "workplace", "relevant", "education", "income", "age", "gender", "gender_txt", "race", "race_txt"]
```

```python
survey.columns = col_names
```

```python
survey.head()
```

Check for missing data

Drop all rows where all values are NA. There are no columns where all values are NA.

```python
survey = survey.dropna(how='all')
```

Drop all rows that contain less than five answers.

```python
survey = survey.dropna(thresh=5)
```

```python
survey.head()
```

Check datatypes

```python
survey.dtypes
```

Check for dupliate rows (may be same respondent submitting twice)

```python
survey["possible_duplicate"]= survey.duplicated()
```

```python
survey
```

```python
survey = survey.drop('possible_duplicate', axis=1)
survey.head()
```

Remove "Cissexual antigender" (49), "Human" (76), and "White latiino"(38) respondent rows

```python
survey = survey.drop([49, 76, 38])
```

Drop text columns for gender and race

```python
survey = survey.drop(['gender_txt', 'race_txt'], axis = 1)
```
